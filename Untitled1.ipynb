```python
print(type(graph))
nodes, edges = ox.graph_to_gdfs(graph)
print(type(nodes))
print(type(edges))

print(nodes.head())
print(nodes.index)
```

```python
print("Graph has:")
G = graph
print(G.number_of_nodes(), "Nodes")
print(G.number_of_edges(), "Edges")
```

```python
edges_to_remove = []

# Find edges that are not highways (roads)
for n in G.edges:
    # print(n)
    data = G.edges[n]
    if('highway' not in data):
        print("Not a highway")
        edges_to_remove.append(n)

# Remove them
for edge_id in edges_to_remove:
    G.remove_edge(edge_id[0], edge_id[1])

# Trim nodes with no connection
remove = [node for node,degree in dict(G.degree()).items() if degree == 0]
G.remove_nodes_from(remove)
```

```python
nodes, edges = ox.graph_to_gdfs(G)
```

```python
# nodes, edges = ox.graph_to_gdfs(G)

node_id = nodes.index[2129]
print(node_id)
# node_id = 13565571

print("Finding node")
selected_node = nodes[nodes.index == node_id]

edge_ids = []
node_ids = []

print("Finding connected nodes")
for id in G.adj[node_id]:
    #print(id)
    node_ids.append(id)
    # edge = G[id][node_id][0]
    # edge_ids.append(edge['osmid'])
    # print(edge)
    
#print(edge_ids)

# connected_edges = edges[edges['osmid'].isin(edge_ids)]

print("Finding connected edges")
connected_edges = edges.loc[node_id, :, :]
connected_nodes = nodes[nodes.index.isin(node_ids)]

#print(connected_nodes)
ax = plot_network()

# edge = edges[edges['osmid'].isin(edges[])]
```

```python
def plot_network():
    fig, ax = plt.subplots(figsize = (24, 24))

    nodes.plot(ax = ax, color = "black", alpha = 0.5, markersize = 1)
    edges.plot(ax = ax, color = "black", alpha = 0.03)

    selected_node.plot(ax = ax,  markersize = 50, color = "green", zorder = 10)

    connected_edges.plot(ax = ax)
    connected_nodes.plot(ax = ax, markersize = 50, color = "orange", zorder = 10)
    
    return ax
```
